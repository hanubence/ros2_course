```python
import numpy
import matplotlib.pyplot as plt

position = numpy.array([0, 4, 0])

print(position)
```

```python
# Create linspace of a circle on the x-y plane with radius r
# with respect to the origin defined by position

r = 1

theta = numpy.linspace(0, 2 * numpy.pi, 100)
x = position[0] + r * numpy.cos(theta)
y = position[1] + r * numpy.sin(theta)

plt.plot(x, y)
```
